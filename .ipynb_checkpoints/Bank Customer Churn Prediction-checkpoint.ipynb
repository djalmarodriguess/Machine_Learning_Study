## Pareto Principle

```python
# Get the value counts of the 'Age' column for churned customers
age_counts = df[df['Exited'] == 1]['Age'].value_counts().reset_index()

# Rename the columns for clarity
age_counts.columns = ['Age', 'Count']

# Calculate the cumulative percentage
age_counts['cumulative_percentage'] = age_counts['Count'].cumsum() / age_counts['Count'].sum() * 100

# Find out which ages contribute to 80% of the churn
pareto_age = age_counts[age_counts['cumulative_percentage'] <= 80]['Age'].sort_values(ascending = True).tolist()

print(f"The ages that contribute to 80% of the churn are: {pareto_age}")
```

```python
plt.figure(figsize=(17,6))
sns.countplot(data=df, x='Age', hue='Exited')
plt.xticks(rotation= 90);
```

According the Pareto principle its that customers between the ages of 33 and 57 are contributing to 80% of the churn. This is a wide age range and it might be beneficial to further analyze this group to understand the specific reasons for the high churn rate.

# Training the Model: Split data into a training set and a test set.

```python
# Split the data into 80% training dataset and 20% test dataset.
from sklearn.model_selection import train_test_split

X = df.iloc[:,:-1]                    # Select Features
y = df['Exited'].copy()               # Select independent variable

#Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.2, random_state = 40 )
```

## Standardize data 

```python
# Standardize training and test datasets.
# It's used to standardize features by removing the mean and scaling to unit variance.
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Model Selection

I am tasked with choosing a suitable machine learning model for prediction. The options at my disposal include Logistic Regression, Decision Trees, Random Forests and Gradient Boosting, which are commonly used models for churn prediction. I will evaluate each model based on its performance and suitability for the data at hand, and select the one that provides the most accurate and reliable predictions.

# Model Evaluation

```python
classifiers = {'Logistic Regression': LogisticRegression(), 'Decision Tree': DecisionTreeClassifier(),
               'Random Forest': RandomForestClassifier(), 'Gradient Boosting': GradientBoostingClassifier()}

for model_name, model in classifiers.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(model_name)
    print(classification_report(y_test, y_pred))
    print('--------------------------------------------------------')
```

# Conclusion

Overall, the Gradient Boosting model is the best choice for classifying customer churn. It has the highest overall accuracy and F1-scores, indicating that it is good at both identifying customers who are likely to churn and predicting which customers will not churn.
