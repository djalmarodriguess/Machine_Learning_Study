```python
from pycaret.internal.pycaret_experiment import TimeSeriesExperiment
from sktime.utils.plotting import plot_series
```

```python
from pycaret.time_series import TSForecastingExperiment
exp = TSForecastingExperiment()
```

```python
exp.setup(df1['TOTAL_QTD'], fh=4, fold=3, session_id=42);
```

```python
melhores_modelos_3 = exp.compare_models(exclude=['auto_arima'], n_select=3, turbo=False)
```

```python
melhores_modelos_3
```

```python
modelo = exp.create_model("gbr_cds_dt")
modelo
```

```python
semanas = 21
y_treino = df1.iloc[:-semanas,0:1].copy()
y_teste = df1.iloc[-semanas:,0:1].copy()
```

```python
y_predito = exp.predict_model(modelo)
```

```python
plt.figure(figsize=(20,6))
plt.plot(y_treino['TOTAL_QTD'], color='b')
plt.plot(y_teste['TOTAL_QTD'], color='orange')

plt.show()
```
