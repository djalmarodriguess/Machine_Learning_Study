```python
# Importing the libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn import metrics
from sklearn import linear_model as lm
```

# Load Data

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
housing
```

```python
print(housing.DESCR)
```

```python
# Initializing the dataframe
df = pd.DataFrame(housing.data)
df.head()
```

```python
housing
```

# Visualizing and Analysis Data

```python
#Adding the feature names to the dataframe
df.columns = housing.feature_names
df.head()
```

```python
#Adding target variable to dataframe
df['Price'] = housing.target
df.head()
```

```python
# Plot histogram for each column of data.

n = len(df.columns)
ncols = 3
nrows = n // ncols 
nrows += n % ncols

fig, axs = plt.subplots(nrows, ncols, figsize=(10, 8))
axs = axs.flatten()

for i, col in enumerate(df.columns):
    df[col].hist(ax=axs[i], bins=50, edgecolor="black")
    most_frequent = df[col].value_counts().idxmax()
    axs[i].set_title(f'{col} (Most frequent: {most_frequent})')

if len(df.columns) < len(axs):
    for idx in range(len(df.columns), len(axs)):
        fig.delaxes(axs[idx])

plt.tight_layout()
plt.show()
```

# Data Preparation

```python
# Features
X_train = df.drop(['Price'], axis = 1)

# Response variable
y_train = df['Price'].copy()
```

# Model Training

```python
# Model description
model= LinearRegression()


# Model Trainig
model.fit(X_train, y_train )
```

```python
# Prediction
pred = model.predict(X_train)
```

```python
pred[0:100]
```

```python
# Add Predict in Data Frame
df['Predict'] = pred
```

```python
df.head(30)
```

```python
df.tail()
```

```python
# Verify the model metrics 

# Calculate the MAE
mae = mean_absolute_error(df['Price'], df['Predict'])
print("Mean Absolute Error:", mae)

# Calculate the MSE
mse = mean_squared_error(df['Price'], df['Predict'])
print("Mean Squared Error:", mse)

# Calculate the R²
r2 = r2_score(df['Price'], df['Predict'])
print("Coefficient of Determination (R²):", r2)
```
